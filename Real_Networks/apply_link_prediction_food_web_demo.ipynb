This script provides an example of how to run the stacking model link prediction code for one food web over one iteration of 5 fold cross-validation. The link prediction script prints out details while it is running about the train and test sets and the features used. The intermediate files and final results files for this demonstration example are all saved in the Results_Demo folder.

```python
import sys
sys.path.append("../Methods/stacking_model") 
import os
import link_prediction_helper
import time
import pickle
import shutil

with open('./Processed_Data_Disaggregated_Lifestage/fw_attr_types.pickle', 'rb') as handle:
    attribute_types = pickle.load(handle)
with open('./Processed_Data_Disaggregated_Lifestage/fw_ids.pickle', 'rb') as handle:
    fw_ids = pickle.load(handle)
```

```python
# Clear out Results_Demo first
folder_path = './Results_Demo'
for item in os.listdir(folder_path):
    item_path = os.path.join(folder_path, item)
    if os.path.isfile(item_path):
        os.remove(item_path)
    elif os.path.isdir(item_path):
        shutil.rmtree(item_path)

curr_web = 'Safford Lake'
fw_id = fw_ids[curr_web]
in_stem = f'Processed_Data_Disaggregated_Lifestage/{curr_web}_{fw_id}/{fw_id}'
temp_stem = fw_id
out_stem = f'{curr_web}_{fw_id}'
details = True # print details to the console while script is running
attr_types = attribute_types[curr_web]
extra_links = False
res_folder = 'Results_Demo'
seed = 55
K = 3
metric = "PR-AUC"

t0 = time.time()
curr_arg = [in_stem,temp_stem,out_stem,details,attr_types,extra_links,res_folder,seed,K,metric]
link_prediction_helper.run_stacking_prediction_comparison(curr_arg)
t1 = time.time()
print("time - " + str(t1-t0))
```
