This file shows an example of running the stacked generalization link prediction approach on one of the synthetic networks. The details parameter is set to true so that intermediate steps in the process can be followed. 

The Results folder for this are saved in Results_Demo as demonstration.

Note this was run with Python 3.12.4

```python
import os
import sys
sys.path.append(os.path.join('..','Methods','stacking_model'))
import link_prediction_helper
import shutil
import time
```

```python
# Clear out Results_Demo first
folder_path = 'Results_Demo'
for item in os.listdir(folder_path):
    item_path = os.path.join(folder_path, item)
    if os.path.isfile(item_path):
        os.remove(item_path)
    elif os.path.isdir(item_path):
        shutil.rmtree(item_path)

curr_web = 'n3_m10_struc' # just pick one as an example.
in_stem = os.path.join('Synth_Nets_Directed_Assortative',curr_web)
temp_stem = curr_web
out_stem = curr_web
details = True
attr_types = [0,0,1,1] # two binary then two numeric 
extra_links = False
res_folder = folder_path
seed = 22
K = 3
opt = 'PR-AUC'

t0 = time.time()
curr_arg = [in_stem,temp_stem,out_stem,details,attr_types,extra_links,res_folder,seed,K,opt]
link_prediction_helper.run_stacking_prediction_comparison(curr_arg)
t1 = time.time()
print(f'time - {str(t1-t0)}')
```
